```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots()

for x in range(5):
    ax.plot(x, x*x, 'o')

# Set axis labels
ax.set_xlabel('X')
ax.set_ylabel('Y')

# Set title
ax.set_title('Quadratic Plot')

# Display the plot
plt.show()
```
